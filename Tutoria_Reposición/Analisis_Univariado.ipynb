```python
#Importamos las librerias necesarias #
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Leer la información
df = pd.read_csv("WA_Fn-UseC_-Telco-Customer-Churn.csv")
```

```python
#Distribución de churn
sns.countplot(x="Churn", data=df, color = "brown")
plt.title("Distribución de Churn")
plt.show()
```

```python
#Distribución del contrato
sns.countplot(x="Contract", data=df, color = "orange")
plt.title("Tipo de contrato")
plt.show()
```

```python
#tenure
#Distribución de meses en el servicio
sns.histplot(df["tenure"], bins=50, kde=True, color="red")
plt.title("Dsitribución de Tenure")
plt.show()
```

```python
sns.boxplot(y=df["tenure"], color="green")
```

```python
sns.histplot(df["MonthlyCharges"], bins=30, kde=True)
```

```python
sns.boxplot(y=df["MonthlyCharges"])
```

```python
#Limpiar Totalcharges
df["TotalCharges"] = pd.to_numeric(df["TotalCharges"].replace(" ", np.nan), errors="coerce")
```

```python
df["TotalCharges"].isna().sum()
```

```python
df["TotalCharges"] = df["TotalCharges"].fillna(0)
```

```python
df["TotalCharges"].describe()
```

```python
#Analisis Bivariado
sns.boxplot(x="Churn", y="TotalCharges", data=df)
plt.title("TotalCharges vs Churn")
plt.show()
```

```python
#Analisis Bivariado
sns.boxplot(x="Churn", y="tenure", data=df)
plt.title("tenure vs Churn")
plt.show()
```
