## Bloque B — IA, Machine Learning y Deep Learning

**¿Son lo mismo?**

El director de ShopSmart dijo *"quiero IA"*. Pero IA, Machine Learning y Deep Learning no son sinónimos — son tres herramientas distintas para problemas distintos. Elegir la incorrecta tiene consecuencias reales en tiempo, costo y resultado.

Piénsalos como tres círculos concéntricos:

```
┌──────────────────────────────────────┐
│  INTELIGENCIA ARTIFICIAL             │
│  Todo sistema que hace algo que      │
│  antes solo hacía un humano          │
│                                      │
│   ┌──────────────────────────────┐   │
│   │  MACHINE LEARNING            │   │
│   │  La máquina aprende de datos │   │
│   │  sin que nadie le escriba    │   │
│   │  las reglas                  │   │
│   │                              │   │
│   │   ┌──────────────────────┐   │   │
│   │   │   DEEP LEARNING      │   │   │
│   │   │   Redes neuronales   │   │   │
│   │   │   con muchas capas   │   │   │
│   │   └──────────────────────┘   │   │
│   └──────────────────────────────┘   │
└──────────────────────────────────────┘
```

| Concepto | En palabras simples |
|----------|--------------------|
| **IA** | Cualquier programa que hace algo que antes solo hacían los humanos |
| **Machine Learning** | Le das ejemplos históricos → el sistema descubre las reglas solo |
| **Deep Learning** | ML con redes neuronales de múltiples capas. Necesita muchos datos. |

**La diferencia clave:**
- Programación tradicional → tú escribes las reglas, la máquina las ejecuta
- Machine Learning → tú das ejemplos, la máquina descubre las reglas

**Para ShopSmart:** predecir si un cliente vuelve es un problema con datos en tabla → **ML clásico**.
Si analizáramos fotos de los productos → Deep Learning. Lo verás en el Módulo 2.

| Problema | Herramienta |
|----------|-------------|
| Predecir retorno de clientes (datos en tabla) | ML clásico (scikit-learn) |
| Clasificar imágenes de productos | Deep Learning |
| Analizar reseñas de clientes en texto | Deep Learning (NLP) |
| Reglas de negocio fijas que no cambian | Programación tradicional |

````python
# ── Celda 2 — Sin ML: el programador inventa las reglas ─────────────────────
def regla_manual(dias_sin_compra, num_compras, cancelaciones):
    #Devuelve 1 -> para vuelve
    #Devuelve 0 -> para no vuelve o probablemente no vuelve
    
    if dias_sin_compra > 180 or cancelaciones >= 3:
        return 0
    
    if num_compras >= 10 and dias_sin_compra < 60:
        return 1
    
    return 0

print("Cliente A", regla_manual(220, 4, 1))
print("Cliente B", regla_manual(30, 12, 0))
print("Cliente C", regla_manual(90, 3, 0))

````

````python
# ── Celda 3 — Con ML: el modelo aprende las reglas de datos históricos ───────
from sklearn.ensemble import RandomForestClassifier
import numpy as np


#X_train: datos de entrada (las features)
# orden: [días_sin_compra, num_compras, gasto_total]

X_train = np.array([
    [12,  2, 120],
    [35,  5, 380],
    [80,  1,  95],
    [15, 10, 920],
    [220, 2, 140],
    [45,  7, 560],
    [300, 1,  60],
    [8,  12, 1100]
])

# y_train: Etiqueta historica
# 1 = volvió, 0 = no volvió
y_train = np.array([1,1,0,1,0,1,0,1])

#Crear el modelo
#Random_state = fijar la aleatoriedad para reproducir resultados
modelo = RandomForestClassifier(n_estimators=20, random_state=42)


#Entrenamiento, patrones (X_train, y_train)
modelo.fit(X_train, y_train)

#Nuevos clientes
clientes_nuevos = np.array([
    [18,  9, 780],
    [260, 2, 130],
    [55,  6, 420],
    [7,  14, 1250]
])

#prediccion
predicciones = modelo.predict(clientes_nuevos)

#probabilidad
probabilidades = modelo.predict_proba(clientes_nuevos)[:, 1]

for i, (p, proba) in enumerate(zip(predicciones, probabilidades), start=1):
    estado = "Va a volver" if p == 1 else "No va a volver"
    print(f"Cliente nuevo {i}: {estado} | probabilidades de volver = {proba:.2f}")




````

## Bloque C — Ética y Sesgos en IA

> *Un modelo aprende lo que los datos dicen — incluyendo sus errores históricos.*

El modelo de la Celda 3 aprendió de ejemplos pasados. Eso es poderoso. Pero tiene un riesgo que pocas veces se menciona: **si los datos históricos tienen sesgos, el modelo los aprende, los amplifica y los convierte en decisiones automáticas a escala.**

**Tres casos reales:**

| Caso | Qué pasó | Por qué |
|------|----------|--------|
| Amazon CVs (2018) | La IA penalizaba la palabra "mujeres" en los CVs | Fue entrenada con historial de contrataciones donde dominaban los hombres |
| Reconocimiento facial MIT (2018) | 0.8% de error en hombres de piel clara vs 34.7% en mujeres de piel oscura | Los datos de entrenamiento no representaban a todos los grupos |
| Apple Card (2019) | Discriminación por código postal | El código postal correlaciona con raza — el modelo discriminó sin que nadie lo programara |

**En ShopSmart:** si clientes de ciertas zonas compran menos, ¿es porque no quieren comprar? ¿O porque los costos de envío son más altos para esas zonas? Si el modelo aprende ese patrón y dejamos de enviarles ofertas, perpetuamos una inequidad con cada predicción automática.

**Los 3 principios — en cada proyecto, antes de hacer deploy:**

1. **Representatividad** → ¿mis datos cubren a todos los afectados?
2. **Transparencia** → ¿puedo explicar cada decisión del modelo?
3. **Impacto diferencial** → si el modelo se equivoca, ¿quién paga el precio?

## Bloque D — CRISP-DM: el proceso estándar

**El problema sin proceso:**

El director nos pide el modelo. ¿Por dónde empezamos? ¿Abrimos el CSV y corremos un algoritmo? Sin un orden claro, el 90% de los proyectos de datos falla — no por falta de habilidad técnica, sino por falta de proceso.

**CRISP-DM** *(pronunciado "CRISP Di Em")* es el proceso estándar que usa la industria. Son seis pasos en ciclo — se puede volver a pasos anteriores cuando los resultados lo requieren.

```
  1. Business Understanding  →  ¿Qué necesita el negocio exactamente?
            ↓
  2. Data Understanding      →  ¿Qué datos tenemos? ¿Qué problemas tienen?
            ↓
  3. Data Preparation        →  Limpiar y transformar para el modelo
            ↓
  4. Modeling           ←───   Entrenar y comparar modelos
            ↓        (volver si los resultados no son buenos)
  5. Evaluation              →  ¿El modelo resuelve el problema real?
            ↓
  6. Deployment              →  Poner en producción
```

| Paso | En ShopSmart | Cuándo |
|------|-------------|--------|
| 1. Business Understanding | Definir qué significa "cliente en riesgo" | **Hoy** |
| 2. Data Understanding | Explorar y diagnosticar el CSV | **Hoy** |
| 3. Data Preparation | Limpiar nulos, fechas, encoding | Clase 2 |
| 4. Modeling | Entrenar Random Forest, comparar modelos | Módulo 2 |
| 5. Evaluation | Precisión, Recall, F1-Score | Módulo 2 |
| 6. Deployment | API que genera lista de riesgo cada lunes | Módulo 3 |

> *"El CRISP-DM no es burocracia. Es la diferencia entre construir modelos útiles y construir modelos que nadie usa."*

## Bloque E — Primer contacto con NumPy

**La pregunta que justifica NumPy:**

El director quiere que el modelo analice a sus clientes. Ya cargamos el CSV y vimos sus datos. Pero hay un problema: el CSV tiene texto (`"Ropa"`), fechas (`"2/12/2023"`), identificadores (`"CLI-0001"`). Un modelo de ML no puede procesar ninguna de esas cosas directamente.

**Todo modelo de IA, sin excepción, trabaja con matrices de números.**

Texto, imágenes, fechas, categorías — todo se convierte a números antes de entrar al modelo. NumPy es la librería que hace esa conversión posible y eficiente.

**¿Por qué no usar listas de Python?**

```python
# Lista Python: multiplicar por 2 DUPLICA la lista
[100, 200, 300] * 2  →  [100, 200, 300, 100, 200, 300]

# NumPy: multiplicar por 2 actúa sobre CADA elemento
np.array([100, 200, 300]) * 2  →  [200, 400, 600]
```

NumPy es entre **50 y 200 veces más rápido** para operaciones matemáticas masivas. Con 200 clientes la diferencia es pequeña. Con miles de iteraciones de entrenamiento sobre el dataset completo, la diferencia se convierte en horas.
